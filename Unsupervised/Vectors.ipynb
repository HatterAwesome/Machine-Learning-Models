```python
documents = [
    "The cat sat on the mat.",
    "The dog sat on the log.",
    "Cats and dogs are pets."
]
```

```python
from sklearn.preprocessing import OneHotEncoder

words = [word.lower().strip(string.punctuation)
         for doc in documents for word in doc.split()]

vocabulary = sorted(set(words))

encoder = OneHotEncoder(sparse_output=False)
one_hot_vectors = encoder.fit_transform(np.array(vocabulary).reshape(-1, 1))

word_to_onehot = {
    vocabulary[i]: one_hot_vectors[i]
    for i in range(len(vocabulary))
}

for word, vector in word_to_onehot.items():
    print(f"{word}: {vector}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer()
X_bow = vectorizer.fit_transform(documents)

print("BoW Matrix:\n", X_bow.toarray())
print("Features:\n", vectorizer.get_feature_names_out())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
X_tfidf = tfidf_vectorizer.fit_transform(documents)

print("TF-IDF Matrix:\n", X_tfidf.toarray())
print("Features:\n", tfidf_vectorizer.get_feature_names_out())
```

```python
count_vectorizer = CountVectorizer()
X_count = count_vectorizer.fit_transform(documents)

print("Count Vectorizer Matrix:\n", X_count.toarray())
print("Features:\n", count_vectorizer.get_feature_names_out())
```

```python
# Download small GloVe embeddings
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip glove.6B.zip

# Load embeddings
embeddings_index = {}
with open("glove.6B.50d.txt", encoding="utf8") as f:
    for line in f:
        values = line.split()
        word = values[0]
        vector = np.asarray(values[1:], dtype='float32')
        embeddings_index[word] = vector

# Example
print("Vector for 'cat':\n", embeddings_index.get("cat"))
```
